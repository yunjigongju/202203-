```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline 
sns.set_style("whitegrid")
iris = sns.load_dataset("iris") 
x = iris.petal_length.values
sns.displot(x, rug=True)
plt.show()
```

```python
titanic = sns.load_dataset("titanic")
sns.countplot(x = "class", data=titanic)
plt.show()
```

```python
tips = sns.load_dataset("tips")
print(tips)
sns.countplot(x="day", data=tips)
plt.show()
```

```python
#   
tips.head(5)
```

```python
#    boxplot 
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
#   ,   
sns.boxplot(x="day", y="total_bill", hue = "smoker" ,data=tips)
plt.show()
```

```python
#   ,  
sns.boxplot(x="day", y="total_bill", hue = "sex" ,data=tips)
plt.show()
```

```python
#   
sns.violinplot(x="day", y="total_bill", hue = "sex" ,data=tips)
plt.show()
```

```python
# bar 
sns.barplot(x="day", y="total_bill", hue = "sex" ,data=tips)
plt.show()
```

```python
# lmplot 
sns.set_style("white") # 
sns.lmplot(x = "total_bill", y = "tip", data=tips, size=7)
plt.show()
```

```python
#     ,  (Set1~3)
sns.lmplot(x = "total_bill", y = "tip", hue = "smoker", data=tips, palette = "Set2", size=7)
plt.show()
```

```python
flights = sns.load_dataset("flights")
type(flights)
#  pivot(, , )
fgp=flights.pivot("month", "year", "passengers") 
plt.figure(figsize=(10,8))
# YlGnBu   
sns.heatmap(fgp, cmap="YlGnBu", annot = True, fmt = 'd')
plt.show()
```

```python
flights = sns.load_dataset("flights")
type(flights)
fgp=flights.pivot("month", "year", "passengers")
plt.figure(figsize=(10,8))
# YlGnBu   
# RdYlGn_r
sns.heatmap(fgp, cmap="RdYlGn_r", annot = True, fmt = 'd')
plt.show()
```

```python
# stripplot    
#jitter=True       
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True)
plt.show()
```

```python
titanic = sns.load_dataset("titanic")
tit = titanic.pivot_table(index=["sex"], columns=["class"], aggfunc='size')
# linewidths  
sns.heatmap(tit, annot=True, fmt='d', linewidths=20)
plt.show()
```

```python
sns.set_style('ticks')
iris = sns.load_dataset("iris")
iris
sns.pairplot(iris, hue="species")
plt.show()
```
