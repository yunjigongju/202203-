```python
fig = plt.figure(constrained_layout = True)
fig.add_subplot()
#그래프 위치지정
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
#원하는 위치에 그래프 작성
x = range(0, 10)
y = range(5, 15)
ax1.plot(x,y,'ro')
c = range(5,50)
d = range(5,50)
ax2.plot(c,d,'ms')
#x축, y축 이름 지정
ax1.set_xlabel('time')
ax1.set_ylabel('qty')
ax2.set_xlabel('time')
ax2.set_ylabel('qty')
plt.show()
```

```python
import matplotlib.font_manager as fm
import matplotlib
#간격을 가능한 최적의 수치로 자동 조정
fig = plt.figure(constrained_layout = True)
#폰트지정
font_location = "C:\Windows\Fonts\H2PORM.TTF"
fontname = fm.FontProperties(fname=font_location).get_name()
matplotlib.rc('font',family = fontname)
fig.add_subplot()
#그래프 위치지정
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
#원하는 위치에 그래프 작성
x = range(0, 10)
y = range(5, 15)
ax1.plot(x,y,'ro')
c = range(5,50)
d = range(5,50)
ax2.plot(c,d,'ms')
#x축, y축 이름 지정
ax1.set_xlabel('시간')
ax1.set_ylabel('qty')
ax2.set_xlabel('time')
ax2.set_ylabel('qty')
plt.show()
```

```python
import matplotlib.pyplot as plt
#막대그래프
plt.style.use('ggplot')
fig = plt.figure()
ax = fig.add_subplot(111)
x = [3, 4]
y = [1, 2]
# x축 항목이름 지정
labels = ['english', 'math']
# 색깔 지정
colors = ['red', 'blue']
# bar() 막대그래프 그리는 함수
ax.bar(x,y, tick_label = labels, color = colors)
for i, v in enumerate(y) :
    ax.text(v+2, i+1.05, str(v), color = 'red', fontweight = 'bold')
plt.show()
```

```python
plt.style.use('ggplot')
fig = plt.figure()
ax = fig.add_subplot(111)
labels = ['english', 'math']
x = [3,4]
y = [1,2]
ax.barh(x,y,tick_label = labels)
plt.show()
```

```python
#묶음 막대그래프 작성
import numpy as np
x = [1,2]
y1,y2,y3 = [1,2],[2,5],[3,7]
fig = plt.figure()
ax = fig.add_subplot(111)
w = 0.2
ax.bar(x,y1,width = w, label = '영어')
ax.bar(np.array(x)+w,y2,width = w, label = '수학')
ax.bar(np.array(x)+w*2,y3,width = w, label = '과학')
#막대그래프 쪽에 뜨는 범례
ax.legend()
plt.show()
```

```python
#pie차트 만들기(원차트)
plt.style.use('ggplot')
fig = plt.figure()
ax = fig.add_subplot(111)
label = ['영어', '수학', '과학']
sizes = [45,35,20]
ax.pie(sizes, labels = label)
plt.show()
```

```python
#수치입력하기

plt.style.use('ggplot')
fig = plt.figure()
ax = fig.add_subplot(111)
label = ['영어', '수학', '과학']
sizes = [45,35,20]
# %를 문자로 입력하기위해 두번씀, 1.1은 자릿수 지정한것 / 그림자 옵션 (옵션은 항상 뒤에추가)
ax.pie(sizes, labels = label, autopct='%1.1f%%', shadow=True)
#차트제목 설정
ax.set_title('우리반 시험 점수 차트')
plt.show()
```

```python
# box plot 그리기
plt.style.use('ggplot')
fig = plt.figure()
ax = fig.add_subplot(111)
x = [1,2,3,4,11,20]
ax.boxplot(x)
plt.show()
```

```python
#여러개의 boxplot
plt.style.use('ggplot')
fig = plt.figure()
ax = fig.add_subplot(111)
x = [[1,2,3,4,11,20],[2,5,15,30,35,40]]
label = ['english','math']
ax.boxplot(x, labels = label)
plt.show()
```
