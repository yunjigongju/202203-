```python
import matplotlib.pyplot as plt
#
plt.style.use('ggplot')
fig = plt.figure()
ax = fig.add_subplot(111)
x = [3, 4]
y = [1, 2]
# x  
labels = ['english', 'math']
#  
colors = ['red', 'blue']
# bar()   
ax.bar(x,y, tick_label = labels, color = colors)
for i, v in enumerate(y) :
    ax.text(v+2, i+1.05, str(v), color = 'red', fontweight = 'bold')
plt.show()
```

```python
plt.style.use('ggplot')
fig = plt.figure()
ax = fig.add_subplot(111)
labels = ['english', 'math']
x = [3,4]
y = [1,2]
ax.barh(x,y,tick_label = labels)
plt.show()
```

```python
#  
import numpy as np
x = [1,2]
y1,y2,y3 = [1,2],[2,5],[3,7]
fig = plt.figure()
ax = fig.add_subplot(111)
w = 0.2
ax.bar(x,y1,width = w, label = '')
ax.bar(np.array(x)+w,y2,width = w, label = '')
ax.bar(np.array(x)+w*2,y3,width = w, label = '')
#   
ax.legend()
plt.show()
```

```python
#pie ()
plt.style.use('ggplot')
fig = plt.figure()
ax = fig.add_subplot(111)
label = ['', '', '']
sizes = [45,35,20]
ax.pie(sizes, labels = label)
plt.show()
```

```python
#

plt.style.use('ggplot')
fig = plt.figure()
ax = fig.add_subplot(111)
label = ['', '', '']
sizes = [45,35,20]
# %   , 1.1   /   (  )
ax.pie(sizes, labels = label, autopct='%1.1f%%', shadow=True)
# 
ax.set_title('   ')
plt.show()
```

```python
# box plot 
plt.style.use('ggplot')
fig = plt.figure()
ax = fig.add_subplot(111)
x = [1,2,3,4,11,20]
ax.boxplot(x)
plt.show()
```
